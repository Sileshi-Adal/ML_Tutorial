```python
from owslib.wms import WebMapService
wms = WebMapService('https://kartta.hel.fi/ws/geoserver/avoindata/wms', version='1.1.1')
```

```python
list(wms.contents)
```

```python
layer='avoindata:Ortoilmakuva_2019_5cm'

print(wms[layer].boundingBox)
print(wms[layer].boundingBoxWGS84)
print(wms[layer].crsOptions)

img = wms.getmap(layers=['avoindata:Ortoilmakuva_2019_5cm'],srs='EPSG:4326',bbox=(24.931,60.162,24.94,60.164),size=(500,200),format='image/jpeg',transparent=True)
```

```python

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from PIL import Image
import datetime
import numpy as np
import matplotlib.pyplot as plt

out = open('out.jpg', 'wb')
out.write(img.read())
out.close()
plt.imshow(mpimg.imread('out.jpg'))
```
