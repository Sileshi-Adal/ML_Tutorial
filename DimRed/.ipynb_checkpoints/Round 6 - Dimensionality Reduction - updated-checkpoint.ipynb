# Machine Learning with Python - Dimensionality Reduction

## Learning goals

In this exercise you will learn how to transform high-dimensional data points (long feature vectors) to shorter feature vectors which retain most of the relevant information. The benefit of such dimensional reduction is two-fold.  Shorter feature vectors typically imply less computation. Moreover, shorter (but relevant) feature vectors ensure good generalization of ML methods. In contrast, using very long feature vectors bears the risk of overfitting. 

We will show-case two particular methods for dimensionality reduction. The first is PCA which amounts to finding or learning a good linear tranformation of the raw feature vectors. The second methos is based on selecting the most relevant features and can be interpreted as a special case of a linear tranfmation using a masking matrix. 

use **principal component analysis (PCA)** for **dimensionality reduction**. You will apply PCA to compress the information contained in (extremely) long raw feature vectors into a small set of new features, which are called **principal components** (PCs). These new features can then be used as the input to regression or classification methods. 

Using a small number of PCs as features is beneficial computationally and statistically. Indeed, using shorter feature vectors typically means less computation. Moreover, replacing long data vectors with short feature vectors reduces the risk that supervised methods overfit the training data. Finally, using feature vectors of length 2 (or 3) allows to visualize high-dimensional data points conveniently using a scatter plot.

## 2.1 Implement PCA
<a id="Q1"></a>

Each fruit image is represented by the raw data vector  $\mathbf{z}^{(i)} \in \mathbb{R}^{D}$, which contains the grayscale values for each of the $D=50^{2}=2500$ pixels. The aim of PCA is to find an (in a certain sense) optimal compression matrix $\mathbf{W} \in \mathbb{R}^{n \times D}$, with $n \ll D$, that transforms the (long) data vector $\mathbf{z}^{(i)} \in \mathbb{R}^{D}$ to a short feature vector $\mathbf{x}^{(i)} \in \mathbb{R}^{n}$ via

<a id="eq1"></a>
\begin{equation}
    \mathbf{x}^{(i)}=\mathbf{W} \mathbf{z}^{(i)}. 
    \tag{1} 
    \label{eq:1}
\end{equation}

We can quantify the loss of information incurred by replacing $\mathbf{z}^{(i)}$ with $\mathbf{x}^{(i)}$ by the smallest possible error achieved by reconstructing (in a linear fashion $\mathbf{R} x^{(i)}$ with some matrix $\mathbf{R}$) the original data vector $\mathbf{z}^{(i)}$ from $\mathbf{x}^{(i)}$. The reconstruction error is given by: 

<a id="eq2"></a>
\begin{align}
 \mathcal{E}(\mathbf{W}) & = \min_{\mathbf{R} \in \mathbb{R}^{D \times n}} (1/m) \sum_{i=1}^{m} \| \mathbf{z}^{(i)} - \mathbf{R} \mathbf{x}^{(i)} \|^{2}  \nonumber \\ 
 & \stackrel{\eqref{eq:1}}{=} \min_{\mathbf{R} \in \mathbb{R}^{D \times n}} (1/m) \sum_{i=1}^{m} \| \mathbf{z}^{(i)} - \mathbf{R} \mathbf{W} \mathbf{z}^{(i)} \|^{2}.
 \label{eq:2}
 \tag{2}
\end{align} 

Note that the optimal reconstruction matrix $\mathbf{R}$ (which achieves the minimum in \eqref{eq:2}) depends on the given compression matrix $\mathbf{W}$. However, we are not interested in the optimal reconstruction matrix itself but only in the associated minimum reconstruction error $\mathcal{E}(\mathbf{W})$. 

In order to construct an optimal compression matrix $\mathbf{W}$ which minimizes the reconstruction error $\mathcal{E}(\mathbf{W})$, we use the **eigenvalue decomposition** of the sample covariance matrix

<a id="eq3"></a>
\begin{equation}
\mathbf{Q} =(1/m) \mathbf{Z}^{T} \mathbf{Z} \mbox{ with the centered data matrix } \mathbf{Z}=\big(\mathbf{z}^{(1)}-\widehat{\mathbf{z}},\ldots,\mathbf{z}^{(m)}- \widehat{\mathbf{z}} \big)^{T} \in \mathbb{R}^{m \times D}. 
\label{eq:3}
\tag{3}
\end{equation}

with the empirical mean $\widehat{\mathbf{z}} = (1/m)\sum_{i=1}^{m}\mathbf{z}^{(i)}$. Since the matrix $\mathbf{Q}$ is [positive semi-definite](https://en.wikipedia.org/wiki/Definiteness_of_a_matrix), it has an eigenvalue decomposition of the form 

\begin{equation}
\mathbf{Q} = \big( \mathbf{u}^{(1)},\ldots,\mathbf{u}^{(D)}\big) {\rm diag} \{\lambda^{(1)},\ldots,\lambda^{(D)} \} 
 \big( \mathbf{u}^{(1)},\ldots,\mathbf{u}^{(D)}\big)^{T}.
\end{equation} 

Here, the orthonormal vectors $\mathbf{u}^{(r)}$ are eigenvectors of $\mathbf{Q}$ which correspond to the decreasingly ordered eigenvalues $\lambda^{(1)} \geq \lambda^{(2)} \geq \ldots \geq \lambda^{(D)}$ of $\mathbf{Q}$. 

It can be shown that the reconstruction error \eqref{eq:2} is minimal for the compression matrix 

<a id="eq4"></a>
\begin{equation}
\label{eq:4}
\tag{4}
\mathbf{W}_{\rm PCA} = \big( \mathbf{u}^{(1)},\ldots,\mathbf{u}^{(n)} \big)^{T},\in \mathbb{R}^{n \times D}, 
\end{equation}

i.e., $\mathcal{E}(\mathbf{W}_{\rm PCA}) = \min_{\mathbf{W} \in \mathbb{R}^{n \times D}} \mathcal{E}(\mathbf{W})$. 

Note that the rows of $\mathbf{W}_{\rm PCA}$ are the eigenvectors $\mathbf{u}^{(1)},\ldots,\mathbf{u}^{(n)}$ of $\mathbf{Q}$ corresponding to the $n$ largest eigenvalues $\lambda^{(1)},\ldots,\lambda^{(n)}$ of $\mathbf{Q}$. Moreover, the minimum reconstruction error $\mathcal{E}(\mathbf{W}_{\rm PCA})$ is achieved 
for the reconstruction matrix $\mathbf{R} = \mathbf{W}_{\rm PCA}^{T}$ and given in terms of the eigenvalues of $\mathbf{Q}$ as  

<a id="eq5"></a>
\begin{equation}
\label{eq:5}
\tag{5}
\mathcal{E}(\mathbf{W}_{\rm PCA}) = \sum_{r = n+1}^{D} \lambda^{(r)}. 
\end{equation}

```python
###################################################################################################
###################################################################################################
###################################################################################################
###################################################################################################


from sklearn.decomposition import PCA
n = 30 #max 30 components
def compute_pca(X,n):
    pca = PCA(n_components = n,svd_solver = 'auto') #create the object
    pca.fit(X)
    W_pca = pca.components_
    return W_pca

Z_centered = Z - np.mean(Z, axis=0) #normalize
W_pca = compute_pca(Z_centered,min(Z_centered.shape))
print()
```

```python
###################################################################################################
###################################################################################################
###################################################################################################
###################################################################################################


def err_pca(X, n):
    """
    Computes reconstruction error.
    
    :param X: Data to be compressed and reconstructed.
    :param n: number of components, scalar.
    
    :return rec_error: reconstruction error value, scalar.
    """
    
    ### STUDENT TASK ### 
    # Compute the reconstruction error for n PCs
    # rec_error = ...
    ### BEGIN SOLUTION
    pca = PCA(n_components = n)
    X_pca = pca.fit_transform(X)
    X_proj = pca.inverse_transform(X_pca)
    rec_error = np.mean(np.square(X - X_proj))
    ### END SOLUTION
    
    return rec_error
```

```python
from matplotlib.lines import Line2D
import matplotlib.gridspec as gridspec
import warnings
warnings.filterwarnings("ignore")

## Input:
#  Z: Dataset
#  n: number of dimensions
#  m_pics: number of pics per class (Apple, Banana). Min 1, Max 15
def plot_reconstruct(Z,n,m_pics=3):
    # subtract the mean from the data
    Z_mean = np.mean(Z,axis = 0)
    Z = Z - Z_mean
    # x=w*z
    X_pca = np.matmul(W_pca[:n,:],Z[:,:,None])
    # x_reversed=r*x
    X_reversed = np.matmul(W_pca[:n,:].T, X_pca)[:,:,0]
    # Add the subtracted Z_mean to the reconstructed data
    X_reversed = X_reversed + Z_mean
    
    # Setup Figure size that scales with number of images
    fig = plt.figure(figsize = (10,10))
    
    # Setup a (n_pics,2) grid of plots (images)
    gs = gridspec.GridSpec(1, 2*m_pics)
    gs.update(wspace=0.0, hspace=0.0)
    for i in range(m_pics):
        for j in range(0,2):
            # Add a new subplot
            ax = plt.subplot(gs[0,i+j*m_pics])
            # Insert image data into the subplot
            ax.imshow(np.reshape(X_reversed[i+(15*j)],(50,50)),cmap='gray',interpolation='nearest')
            # Remove x- and y-axis from each plot
            ax.set_axis_off()
    
    plt.subplot(gs[0,0]).set_title("Reconstructed images using %d PCs:"%(n), size='large', y=1.08)
    # Render the plot
    plt.show()
    
# The values of PCS n to plot for. You can change these to experiment
num_com=[1,5,50]
for n in num_com:
    # If you want to print different amount of pictures, you can change the value of m_pics. (1-15)
    plot_reconstruct(Z,n,m_pics=3)
```

<div class=" alert alert-warning">

Complete the Python function `X = project_data(W_pca, Z, s)`, which reads in the compression matrix $W_{\rm PCA}$, the centered data matrix $\mathbf{Z}$ (see ([3](#eq3))), and a vector $\mathbf{s}=(s_{1},s_{2})^{T}$ which contains the indices of two PCs (e.g., $[1,2]$). The output of the function is a matrix $\mathbf{X} = \big(\mathbf{x}^{(1)},\ldots,\mathbf{x}^{(m)} \big)^{T} \in \mathbb{R}^{m \times 2}$ whose $i$th row is the feature vector $\mathbf{x}^{(i)} = \big(x_{1}^{(i)},x_{2}^{(i)} \big)^{T}$ obtained by projecting the (centered) data point $\mathbf{z}^{(i)}-\widehat{\mathbf{z}}$ (stored in the $i$th row of $\mathbf{Z}$) on the eigenvectors $\mathbf{u}^{(s_{1})}$ and $\mathbf{u}^{(s_{2})}$ of the sample covariance matrix $\mathbf{Q}$ (see ([3](#eq3))).  
</div>

```python
def project_data(W_pca, Z, selected_pcs):
    """
    Project data on given two PCs.
    
    :param W_pca: the compression matrix, shape=(n, D), numpy array.
    :param Z: centered data, shape=(m, D), numpy array.
    :param selected_pcs: Index of selected PCs, list of length 2 (e.g. [1,2] or [8,9]).
    
    :return X: projected data, shape=(m, 2), numpy array.
    """
    
    ### STUDENT TASK ### 
    # Use the two input PCs in selected_pcs to project the data in 2d and save the values in X
    # X = ...
    ### BEGIN SOLUTION
    X = np.matmul(W_pca[selected_pcs-1,:],Z[:,:,None])[:,:,0]
    ### END SOLUTION
    
    return X 
```

### 3.2.4 Significance of Feature weights

Logistic regression not only provides us with a classifier which allows us to compute a predicted label $\hat{y}$ based on the features $\mathbf{x}$ but also some means to determine the significance (or importance) of individual features. In general, some features will have a more impact on the outcome of classification than others. It can be benificial, both computationally and statistically, to ignore irrelevant features in the classification process.  Indeed, using less features typically implies reduced computational complexity and also reduced risk of overfitting the training data (see Round 4).

Logistic regression allows to estimate the relevance of features $x_{r}$ via the relative magnitude of the corresponding weight $w_{r}$. Let's get the feature weights (coefficients) and analyze them.
